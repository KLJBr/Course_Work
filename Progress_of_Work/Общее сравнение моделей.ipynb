```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
with open('Model_results.xlsx', 'rb') as f:
    df_Valence1 = pd.read_excel(f, sheet_name='Resalts_for_Valence', index_col=0)    
    df_Arousal1 = pd.read_excel(f, sheet_name='Resalts_for_Arousal', index_col=0)
    df_Valence2 = pd.read_excel(f, sheet_name='Results_with_lagging_for_Valence', index_col=0)    
    df_Arousal2 = pd.read_excel(f, sheet_name='Results_with_lagging_for_Arousal', index_col=0)
```

## Сравнение для Valence

```python
df_Valence = df_Valence1.append(df_Valence2)
```

```python
fig, ax = plt.subplots(figsize = (15, 14))
ax = sns.barplot(data = df_Valence, x = 'R2_test', y = df_Valence.index, hue = 'Модель')
for container in ax.containers:
    ax.bar_label(container, size = 12, fmt = '%.2f')
plt.rcParams.update({'font.size': 16})
plt.grid()
plt.legend(fontsize= 16)
plt.title('Модели для Valence')
plt.show()
```

## Сравнение для Arousal

```python
df_Arousal = df_Arousal1.append(df_Arousal2)
```

```python
fig, ax = plt.subplots(figsize = (15, 14))
ax = sns.barplot(data = df_Arousal, x = 'R2_test', y = df_Arousal.index, hue = 'Модель')
for container in ax.containers:
    ax.bar_label(container, size = 12, fmt = '%.2f')
plt.rcParams.update({'font.size': 16})
plt.legend(fontsize= 16)
plt.grid()
plt.title('Models for Arousal')
plt.show()
```
